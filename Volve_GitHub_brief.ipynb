### Load Python requirements

```python
from pandas import DataFrame, read_csv

import altair as alt
#alt.renderers.enable('altair_viewer')


import pandas as pd
import numpy as np


#read the file
#file = r'Volve production data_Monthly_brief_Query2.xlsx'
file = r'Query3.xlsx'
df = pd.read_excel(file,index_col=False)
```

### Data

```python
df.head(20)
```

### Basic Plot

```python
spread = alt.Chart(df).mark_area(opacity=0.3).encode(
    x=alt.X('Date', timeUnit='year'),
    y=alt.Y('ci0(BOPD)', axis=alt.Axis(title='BOPD')),
    y2='ci1(BOPD)',
    color='Well_Name'
).properties(
    title='Volve Field Production',
    width=800
)
lines = alt.Chart(df).mark_line().encode(
    x=alt.X('Date', timeUnit='year'),
    y='mean(BOPD)',
    color='Well_Name'
).properties(
    width=800
)
spread + lines
```

### Map with interactivity from Map on Production and BHP data
#### 2 maps to show interactivity

```python
interval = alt.selection_interval()
base = alt.Chart(df).mark_circle().encode(
    alt.X('X_Bottom:Q',
            scale=alt.Scale(domain=(434000, 438000))
        ),
    alt.Y('Y_Bottom',
            scale=alt.Scale(domain=(6477000, 6480000))),

    #x='XX_Bottom',
    #scale=alt.Scale(zero=False)),
    #size='BOPD',
    size=alt.Size('BOPD:Q', scale=alt.Scale(range=[10, 1000])),

    #label='Well_Name',
    color=alt.condition(interval, 'Well_Name', alt.value('lightgray')),
    tooltip='Well_Name',
).properties(
    title='Volve Field Map Showing Producing Wells',
    width=800,
    height=800,

    selection=interval
)



text = base.mark_text(align='left', dx=5, dy=-5).encode(
text='Well_Name'
)




bhp = alt.Chart(df).mark_circle().encode(
    x='Date',
    y='BHP',
    color='Well_Name:O'
    #color='BOPD:Q'
).properties(
    title='Volve Field - BHP',
    width=800,
    height=180,
    selection=interval
).transform_filter(
    interval
)



hist = alt.Chart(df).mark_bar().encode(
    x='Date',
    y='BOPD',
    color='Well_Name'
).properties(
    title='Volve Field - Oil Production',
    width=800,
    height=180,
    #selection=interval
).transform_filter(
    interval
)

hist2 = alt.Chart(df).mark_bar().encode(
    x='Date',
    y='BWPD',
    color='Well_Name:O'
    #color='BOPD:Q'
).properties(
    title='Volve Field - Water Production',
    width=800,
    height=180,
    #selection=interval
).transform_filter(
    interval
)



#scatter = base.encode(y='YY_Bottom') | base.encode(y='YY_Bottom',size='BWPD:Q')
#scatter = base.encode(y='YY_Bottom',size=alt.Size('BOPD:Q')) | base.encode(y='YY_Bottom',size=alt.Size('BWPD:Q'))
#scatter = base.encode(y='YY_Bottom')









base  & bhp & hist & hist2



```

## Map with interactivity from plots below Map

```python
interval = alt.selection_interval()
base = alt.Chart(df).mark_circle().encode(
    alt.X('X_Bottom:Q',
            scale=alt.Scale(domain=(434000, 438000))
        ),
    alt.Y('Y_Bottom',
            scale=alt.Scale(domain=(6477000, 6480000))),

    #x='XX_Bottom',
    #scale=alt.Scale(zero=False)),
    #size='BOPD',
    size=alt.Size('BOPD:Q', scale=alt.Scale(range=[10, 1000])),

    #label='Well_Name',
    color=alt.condition(interval, 'Well_Name', alt.value('lightgray')),
    tooltip='Well_Name',
).properties(
    title='Volve Field Map Showing Producing Wells Sized by BPOD or BWPD',
    width=800,
    height=800,

    selection=interval
)

text = base.mark_text(
    align='center',
    baseline='middle',
    dx=7
).encode(
    text='Well_Name'
)

bhp = alt.Chart(df).mark_circle().encode(
    x='Date',
    y='BHP',
    color='Well_Name:O'
    #color='BOPD:Q'
).properties(
    title='Volve Field - BHP',
    width=800,
    height=180,
    selection=interval
)



hist = alt.Chart(df).mark_bar().encode(
    x='Date',
    y='BOPD',
    color='Well_Name'
).properties(
    title='Volve Field - Oil Production',
    width=800,
    height=180,
    selection=interval

)

hist2 = alt.Chart(df).mark_bar().encode(
    x='Date',
    y='BWPD',
    color='Well_Name:O'
    #color='BOPD:Q'
).properties(
    title='Volve Field - Water Production',
    width=800,
    height=180,
    selection=interval

)



#scatter = base.encode(y='YY_Bottom') | base.encode(y='YY_Bottom',size='BWPD:Q')
#scatter = base.encode(y='YY_Bottom',size=alt.Size('BOPD:Q')) | base.encode(y='YY_Bottom',size=alt.Size('BWPD:Q'))
#scatter = base.encode(y='YY_Bottom')





base + text



base & bhp & hist & hist2

```

# Bubble Plot of Production

```python
interval = alt.selection_interval()
base = alt.Chart(df).mark_circle().encode(
    alt.X('X_Bottom:Q',
            scale=alt.Scale(domain=(434000, 438000))
        ),
    alt.Y('Y_Bottom',
            scale=alt.Scale(domain=(6477000, 6480000))),

    #x='XX_Bottom',
    #scale=alt.Scale(zero=False)),
    #size='BOPD',
    size=alt.Size('BOPD:Q', scale=alt.Scale(range=[10, 1000])),

    #label='Well_Name',
    color=alt.condition(interval, 'Well_Name', alt.value('lightgray')),
    tooltip='Well_Name',
).properties(
    title='Volve Field Map Showing Producing Wells Sized by BPOD or BWPD',
    width=600,
    height=600,

    selection=interval
)


scatter =  base.encode(x='Date', y='BOPD',size='BOPD:Q')
#scatter = base.encode(y='YY_Bottom',size=alt.Size('BOPD:Q')) | base.encode(y='YY_Bottom',size=alt.Size('BWPD:Q'))
#scatter = base.encode(y='YY_Bottom')


base & scatter
```

### Better layout of all Plots

```python
interval = alt.selection_interval()

base = alt.Chart(df).properties(
    width=350,
    height=350,
).add_selection(interval)


points = base.mark_point(filled=True, size=200).encode(

    alt.X('X_Bottom:Q',
            scale=alt.Scale(domain=(434000, 438000))
        ),
    alt.Y('Y_Bottom',
            scale=alt.Scale(domain=(6477000, 6480000))),

    #x='XX_Bottom:Q',
    #y='YY_Bottom:Q',
    size='BOPD:Q',
    color=alt.condition(interval, 'Well_Name', alt.value('lightgray')),
    #color=alt.condition(selector, 'Well_Name:O', alt.value('lightgray'), legend=None),
    tooltip='Well_Name',
).properties(
    title='Volve Field Map',
    selection=interval
)


timeseries = base.mark_line().encode(
    x='Date',
    y=alt.Y('BOPD', scale=alt.Scale(domain=(0, 40000))),
    color=alt.Color('Well_Name:O')

).properties(
    title='Volve Production by Well in BOPD',
    width=350,
    height=140
    #selection=interval
).transform_filter(
    #title='Volve Field BOPD',
    interval
)


timeseries2 = base.mark_line().encode(
    x='Date',
    y=alt.Y('BWPD', scale=alt.Scale(domain=(0, 40000))),
    color=alt.Color('Well_Name:O')

).properties(
    title='Volve Production by Well in BWPD',
    width=350,
    height=140
    #selection=interval
).transform_filter(
    #title='Volve Field BOPD',
    interval
)



bhp = alt.Chart(df).mark_circle().encode(
    x='Date',
    y='BHP',
    color='Well_Name:O'
    #color='BOPD:Q'
).properties(
    title='Volve Field - BHP',
    width=700,
    height=180,
    selection=interval
).transform_filter(
    interval
)


hist = alt.Chart(df).mark_bar().encode(
    #x='count()',
    x='sum(BOPM)',
    y='Well_Name',
    color='Well_Name'
).properties(
    width=700,
    height=80
).transform_filter(
    interval
)

hist2 = alt.Chart(df).mark_bar().encode(
    #x='count()',
    x='sum(BWPM)',
    y='Well_Name',
    color='Well_Name'
).properties(
    width=700,
    height=80
).transform_filter(
    interval
)

hist3 = alt.Chart(df).mark_bar().encode(
    x='sum(beGpm)',
    y='Well_Name',
    color='Well_Name'
).properties(
    width=700,
    height=80
).transform_filter(
    interval
)

first = points | timeseries & timeseries2
second = bhp & hist & hist2 & hist3



#points & timeseries & timeseries2 & bhp & hist & hist2 & hist3

first & second
```

# Cum Data by Well

```python
interval = alt.selection_interval()

base = alt.Chart(df).properties(
    width=300,
    height=300,
).add_selection(interval)

points = base.mark_point(filled=True, size=200).encode(
    alt.X('X_Bottom:Q',
            scale=alt.Scale(domain=(434000, 438000))
        ),
    alt.Y('Y_Bottom',
            scale=alt.Scale(domain=(6477000, 6480000))),

    #x='XX_Bottom:Q',
    #y='YY_Bottom:Q',
    size='BOPD:Q',
    color=alt.condition(interval, 'Well_Name', alt.value('lightgray')),
    #color=alt.condition(selector, 'Well_Name:O', alt.value('lightgray'), legend=None),
    tooltip='Well_Name',
).properties(
    title='Volve Field Map',
    selection=interval


)

timeseries = base.mark_line().encode(
    x='Date',
    y=alt.Y('Cum_Oil', scale=alt.Scale(domain=(0, 30000000))),
    color=alt.Color('Well_Name:O')

).properties(
    title='Volve Cum_Oil by Well',
    #selection=interval


).transform_filter(
    #title='Volve Field BOPD',
    interval
)

timeseries2 = base.mark_line().encode(
    x='Date',
    y=alt.Y('Cum_Water', scale=alt.Scale(domain=(0, 50000000))),
    color=alt.Color('Well_Name:O')

).properties(
    title='Volve Cum_Water by Well',
    #selection=interval


).transform_filter(
    #title='Volve Field BOPD',
    interval
)

timeseries3 = base.mark_line().encode(
    x='Cum_Water',
    y=alt.Y('Cum_Oil', scale=alt.Scale(domain=(0, 30000000))),
    color=alt.Color('Well_Name:O')

).properties(
    title='Volve Cum_Oil and Cum_Water by Well',
    #selection=interval


).transform_filter(
    #title='Volve Field BOPD',
    interval
)


first = points | timeseries
second = timeseries2 | timeseries3


#points & timeseries & timeseries2 & timeseries3
first & second
```
